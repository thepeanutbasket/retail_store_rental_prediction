```python
# 주차가능여부, 승강기여부 변수 추가
lst = []
[lst.append(1) if i > 0 else lst.append(0) for i in seoul_ua['주차대수']]

seoul_ua['주차가능여부'] = lst

lst = []
[lst.append(1) if i > 0 else lst.append(0) for i in seoul_ua['승강기수']]

seoul_ua['승강기여부'] = lst

seoul_ua
```

```python
# 0값 제거
seoul_ua1 = seoul_ua.copy()

seoul_ua1 = seoul_ua1[seoul_ua1.rent_ua != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.지상총층 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.연면적m2 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.usage_area != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.cont_area != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.지하철거리m != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.승객수점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.건물별전기사용량 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.본번별평균사용량 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.횡단보도거리m != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.가까운아파트총세대수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.가까운아파트평균평당가 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.아파트거리 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.주변세대수점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.주변평당가점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균월매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균월매출건수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균주중매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균주말매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.공시지가 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.행정동생활인구 != 0]
seoul_ua1
```

```python
seoul_ua2_log = seoul_ua1.copy()

q1 = seoul_ua2_log["rent_ua"].quantile(0.25)
q3 = seoul_ua2_log["rent_ua"].quantile(0.75)
iqt = 6.0 * (q3 - q1)
seoul_ua2_log["rent_ua"][(seoul_ua2_log["rent_ua"] > (q3 + iqt)) | (seoul_ua2_log["rent_ua"] < (q1 - iqt))] = None

seoul_ua2_log = seoul_ua2_log.dropna(axis = 0, how= 'any')

seoul_ua2_log
```

```python
columns = ['rent_ua', '연면적m2', 'usage_area',
       'cont_area', '지하철거리m', '승객수점수', '건물별전기사용량', '본번별평균사용량', '횡단보도거리m',
       '건물연차', '가까운아파트총세대수', '가까운아파트평균평당가', '아파트거리', '주변세대수점수', '주변평당가점수',
       '평균월매출금액', '평균월매출건수', '평균주중매출금액', '평균주말매출금액', '편의점임대료', '공시지가',
       '행정동생활인구','총유동인구수', '주중평균유동인구수', '주말평균유동인구수',
       '유동인구비율_남성', '유동인구비율_여성', '유동인구비율_10대', '유동인구비율_20대', '유동인구비율_30대',
       '유동인구비율_40대', '유동인구비율_50대', '유동인구비율_60대']


seoul_ua2_log['연면적m2'][seoul_ua2_log['연면적m2'] < seoul_ua2_log['usage_area']] = None
seoul_ua2_log[seoul_ua2_log['연면적m2'].isnull()]

seoul_ua2_log = seoul_ua2_log.dropna(axis = 0, how= 'any')

seoul_ua2_log
```

```python
seoul_ua2_log = seoul_ua2_log.reset_index()
seoul_ua2_log = seoul_ua2_log.iloc[:,1:]

seoul_ua2_log.isnull().sum()
```

```python
# columns에 담긴 변수 log 취하고 scaling
# 도로추정크기에는 0이 들어있으므로 아래에서 다시 실행한다.

import math
from sklearn import preprocessing
import numpy as np

seoul_ua3_log = seoul_ua2_log.copy()

for i in columns:
    seoul_ua3_log['{}_log'.format(i)] = np.log(seoul_ua3_log['{}'.format(i)])
    
seoul_ua3_log.columns
```

```python
seoul_ua3_log = seoul_ua3_log[['rent_ua_log','도로추정크기',
       '연면적m2_log', 'usage_area_log', 'cont_area_log', '지하철거리m_log',
       '승객수점수_log', '건물별전기사용량_log', '본번별평균사용량_log', '횡단보도거리m_log', '건물연차_log',
       '가까운아파트총세대수_log', '가까운아파트평균평당가_log', '아파트거리_log', '주변세대수점수_log',
       '주변평당가점수_log',  '평균월매출건수_log', 
       '평균주말매출금액_log', '편의점임대료_log', '공시지가_log', '행정동생활인구_log', '총유동인구수_log',
       '주중평균유동인구수_log', '주말평균유동인구수_log', '유동인구비율_남성_log', '유동인구비율_여성_log',
       '유동인구비율_10대_log', '유동인구비율_20대_log', '유동인구비율_30대_log', '유동인구비율_40대_log',
       '유동인구비율_50대_log', '유동인구비율_60대_log']]

seoul_ua3_log.isnull().sum()
```

```python
# 도로추정크기에 0이 있는 경우는 그대로 0, 아닌 경우만 log 씌우기

lst = []
for i in seoul_ua3_log['도로추정크기']:
    if i == 0:
        lst.append(0)
    else:
        lst.append(np.log(i))
seoul_ua3_log['도로추정크기_log'] = lst

seoul_ua3_log
```
