```python
aa = []
bb = []
cc = []
for i in road['법정동명'].str.split(' '):
    aa.append(i[0])
    bb.append(i[1])
    cc.append(i[2])


road['시도'] = aa
road['시군구'] = bb
road['읍면동'] = cc

road = road[['시군구','읍면동','지번','지목명','지목코드','토지면적','도로접면코드','도로접면','공시지가']]

road.head()
```

```python
'''
도로접면 데이터가 범주형으로 설정되어 있으므로.... 각 도로의 면적으로 값을 바꾼다.
한 면만 접하고 있으면 그 면의 크기(평균)만, 두 면이상 접하고 있으면 그 면들 크기(임의로 두 면만)의 총합

소로각지(소로한면 + 세로(가)): ((10 + 10) + (10 + (0+8)/2))/2
중로각지(중로한면 + 세로(갸)): ((18.5 + 18.5) + (18.5 + (0+8)/2))/2
세로한면(가): (0+8)/2
소로한면: 10
광대로한면: 25
중로한면: 18.5
광대소각(광대로한면 + 소로한면): 25 + 10
세로각지(가)(세로한면 + 세로한면): (0+8)/2 + (0+8)/2
광대세각(광대로한면 + 세로한면): 25 + (0+8)/2
세로한면(불): (0+8)/2
맹지: 0
세로각지(불): (0+8)/2 + (0+8)/2
지정되지않음: 0
'''

aa = []
for i in merge['도로접면']:
    if i == '소로각지':
        aa.append(((10+10) + (10+(0+8)/2))/2)
    elif i == '중로각지':
        aa.append(((18.5+18.5) + (18.5+(0+8)/2))/2)
    elif i == '세로한면(가)':
        aa.append((0+8)/2)
    elif i == '소로한면':
        aa.append(10)
    elif i == '광대로한면':
        aa.append(25)
    elif i == '중로한면':
        aa.append(18.5)
    elif i == '광대소각':
        aa.append(25+10)
    elif i == '세로각지(가)':
        aa.append((0+8)/2 + (0+8)/2)
    elif i == '광대세각':
        aa.append(25 + (0+8)/2)
    elif i == '세로한면(불)':
        aa.append((0+8)/2)
    elif i == '맹지':
        aa.append(0)
    elif i == '세로각지(불)':
        aa.append((0+8)/2 + (0+8)/2)
    else:
        aa.append(0)

merge['도로추정크기'] = aa        
merge
```
