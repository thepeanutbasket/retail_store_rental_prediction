```python
seoul_ua = seoul[['rent_ua',  '연면적m2','usage_area', 
        '승객수점수', '본번별평균사용량', '평균주말매출금액', '공시지가']]
seoul_ua.head()
```

```python
# 0값 제거
seoul_ua1 = seoul_ua.copy()

seoul_ua1 = seoul_ua1[seoul_ua1.rent_ua != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.연면적m2 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.usage_area != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.본번별평균사용량 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.승객수점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균주말매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.공시지가 != 0]
seoul_ua1
```

```python
# 이상값 데이터만 불러오기
# 일반적으로는 iqt = 1.5 * (q3 - q1)로 정의. 하지만 너무 많은 데이터가 이상치로 빠지는 것을 막기 위해 iqt = 1.8 * (q3 - q1)로 정의하였다.

seoul_ua2 = seoul_ua1.copy()

q1 = seoul_ua2["rent_ua"].quantile(0.25)
q3 = seoul_ua2["rent_ua"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df1 = seoul_ua2[(seoul_ua2["rent_ua"] > (q3 + iqt)) | (seoul_ua2["rent_ua"] < (q1 - iqt))]

q1 = seoul_ua2["연면적m2"].quantile(0.25)
q3 = seoul_ua2["연면적m2"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df2 = seoul_ua2[(seoul_ua2["연면적m2"] > (q3 + iqt)) | (seoul_ua2["연면적m2"] < (q1 - iqt))]

q1 = seoul_ua2["usage_area"].quantile(0.25)
q3 = seoul_ua2["usage_area"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df3 = seoul_ua2[(seoul_ua2["usage_area"] > (q3 + iqt)) | (seoul_ua2["usage_area"] < (q1 - iqt))]

q1 = seoul_ua2["승객수점수"].quantile(0.25)
q3 = seoul_ua2["승객수점수"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df4 = seoul_ua2[(seoul_ua2["승객수점수"] > (q3 + iqt)) | (seoul_ua2["승객수점수"] < (q1 - iqt))]

q1 = seoul_ua2["본번별평균사용량"].quantile(0.25)
q3 = seoul_ua2["본번별평균사용량"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df5 = seoul_ua2[(seoul_ua2["본번별평균사용량"] > (q3 + iqt)) | (seoul_ua2["본번별평균사용량"] < (q1 - iqt))]

q1 = seoul_ua2["평균주말매출금액"].quantile(0.25)
q3 = seoul_ua2["평균주말매출금액"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df6 = seoul_ua2[(seoul_ua2["평균주말매출금액"] > (q3 + iqt)) | (seoul_ua2["평균주말매출금액"] < (q1 - iqt))]

q1 = seoul_ua2["공시지가"].quantile(0.25)
q3 = seoul_ua2["공시지가"].quantile(0.75)
iqt = 1.8 * (q3 - q1)
df7 = seoul_ua2[(seoul_ua2["공시지가"] > (q3 + iqt)) | (seoul_ua2["공시지가"] < (q1 - iqt))]


df8 = pd.concat([df1,df2,df3,df4,df5,df6,df7])

df8
```

```python
df8.duplicated().value_counts()
```

```python
df8 = df8.drop_duplicates()
df8
```

```python
seoul_ua2_log = df8.copy()
seoul_ua2_log
```

```python
# 요약 통계량 확인하기

seoul_ua2_log = seoul_ua2_log[seoul_ua2_log['연면적m2'] != 0]
seoul_ua2_log.describe()
```

```python
seoul_ua2_log.columns
```

```python
columns = ['rent_ua', '연면적m2', 'usage_area', '승객수점수', '본번별평균사용량', '평균주말매출금액',
       '공시지가']
```

```python
# columns에 담긴 변수 log 취하고 scaling

import math
from sklearn import preprocessing

for i in columns:
    seoul_ua2_log['{}_log'.format(i)] = preprocessing.scale(np.log(seoul_ua2_log['{}'.format(i)]))
    
seoul_ua2_log
```

```python
seoul_ua2_log.columns
```

```python
# 로그 없이

# 회귀분석 모듈 import

from statsmodels.formula.api import ols

# 모형 만들기
model = ols('rent_ua ~ 연면적m2 + usage_area + 승객수점수 + 본번별평균사용량 + 평균주말매출금액 + 공시지가', seoul_ua2_log)

# 모형 적합
res = model.fit()

# 분석 결과 확인
res.summary()
```

```python
# 다중공선성 확인을 위한 모듈

from statsmodels.stats.outliers_influence import variance_inflation_factor

# 한번에 모든 컬럼의 VIF를 계산

pd.DataFrame({'컬럼': column, 'VIF': variance_inflation_factor(model.exog, i)} 
             for i, column in enumerate(model.exog_names)
             if column != 'Intercept')  # 절편의 VIF는 구하지 않는다.
```
