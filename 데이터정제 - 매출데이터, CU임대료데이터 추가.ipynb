```python
# geocoding 변환 파일 불러오기
geo_table = pd.read_csv("C:/data/now_used/geo_table.csv")

geo_table = geo_table[['상권코드명','X','Y']]
geo_table
```

```python
# geo_table + money merge
money_geo = pd.merge(money, geo_table, on = '상권코드명', how = 'left')
money_geo.columns
```

```python
# 상가와 가장 가까운 위/경도값을 가진 상권코드명의 평균월매출금액, 평균월매출건수, 평균주중매출금액, 평균주말매출금액 구하기
sanga_seoul = pd.read_csv("C:/data/now_used/real_final_data_all.csv")


import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier


clf = KNeighborsClassifier(n_neighbors = 1)       # 3개의 평균
clf.fit(money_geo[['X','Y']],money_geo['상권코드명'])

x = sanga_seoul['경도'].tolist()
y = sanga_seoul['위도'].tolist()
z = []

for i,j in zip(x,y):
    z.append(clf.predict(np.array([[i,j]]))[0])

z
```

```python
sanga_seoul['상권코드명'] = z

sanga_seoul = pd.merge(sanga_seoul, money_geo[['상권코드명', '평균월매출금액', '평균월매출건수', '평균주중매출금액', '평균주말매출금액']], on = '상권코드명', how = 'left')

sanga_seoul.to_csv("C:/data/now_used/final_data_with_money.csv", index = False)

sanga_seoul
```
