## 상관계수 히트맵 그려보기

```python
# 상관계수 확인

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rcParams['font.family'] = 'NanumGothic'

# 스피어만 상관계수

corr_ua_s = anal_df.corr(method = 'spearman')
corr_ua_s
```

```python
import matplotlib.pylab as plt
from matplotlib import pyplot
from matplotlib import font_manager, rc
import numpy as np
font_name = font_manager.FontProperties(fname = 'C:/windows/fonts/malgun.ttf').get_name()
rc('font', family = font_name)
```

```python
# 스피어만 상관계수 히트맵

mask = np.zeros_like(corr_ua_s, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots( figsize=(20,20) )

sns.heatmap(corr_ua_s,
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = -1,vmax = 1   # 컬러바 범위 -1 ~ 1
           )  
plt.show()
```

```python
# 회귀분석

from statsmodels.formula.api import ols

# 모형 만들기 (전체 변수를 넣어보자)

model1 = ols('rent_ua_log ~ 연면적m2_log + usage_area_log+ cont_area_log+ 지하철거리m_log+ 승객수점수_log+ 건물별전기사용량_log+ 본번별평균사용량_log+ 횡단보도거리m_log+ 건물연차_log+ 가까운아파트총세대수_log+ 가까운아파트평균평당가_log+ 아파트거리_log+ 주변세대수점수_log+ 주변평당가점수_log+ 평균월매출금액_log+ 평균월매출건수_log+ 평균주중매출금액_log+ 평균주말매출금액_log+ 편의점임대료_log+ 공시지가_log+ 행정동생활인구_log+ 주차가능여부+ 승강기여부', anal_df)

# 모형 적합

res = model1.fit()

# 분석 결과 확인

res.summary()
```

## 다중공선성 파악
다중공선성은 분산팽창계수(VIF, Variance Inflation Factor)를 구하여 판단한다. 엄밀한 기준은 없으나 일반적으로 10보다 크면 다중공선성이 있다고 판단. (5를 기준으로 하기도 한다.)

```python
# 다중공선성 확인을 위한 모듈

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# model3(모든 연속형 변수들의 이상값 제거 후 모든 변수를 넣은 모형)의 독립변수 확인

model1.exog_names
```

```python
pd.options.display.float_format = '{:.5f}'.format
```

```python
# 한번에 모든 컬럼의 VIF를 계산

pd.DataFrame({'컬럼': column, 'VIF': variance_inflation_factor(model1.exog, i)} 
             for i, column in enumerate(model1.exog_names)
             if column != 'Intercept')  # 절편의 VIF는 구하지 않는다.
```

## 다중공선성 10 이상 변수 제외, p-value > 0.05 변수 제외

```python
model3 = ols('rent_ua_log ~ 공시지가_log + 평균주말매출금액_log + usage_area_log + 승객수점수_log + 본번별평균사용량_log', anal_df)

# 모형 적합

res = model3.fit()

# 분석 결과 확인

res.summary()
```

```python
# 한번에 모든 컬럼의 VIF를 계산

pd.DataFrame({'컬럼': column, 'VIF': variance_inflation_factor(model3.exog, i)} 
             for i, column in enumerate(model3.exog_names)
             if column != 'Intercept')  # 절편의 VIF는 구하지 않는다.
```

## KNN분석 - Regressor

```python
anal_df.columns
```

```python
df = anal_df[['공시지가_log','평균주말매출금액_log','usage_area_log','승객수점수_log','본번별평균사용량_log']]
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split 

X = df.iloc[:,1:]     # train/test data로 나눌 수 있게 내가 필요한 data만 담기
Y = df.iloc[:,0]      # label 만들기

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
rgs = KNeighborsRegressor(n_neighbors = 3)    # knn class 인스턴스화
rgs.fit(X_train, Y_train)
rgs.predict(X_test)
```

```python
rgs.score(X_train,Y_train)
```

```python
rgs.score(X_test,Y_test)
```
