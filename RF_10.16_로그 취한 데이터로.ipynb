```python
seoul = pd.read_csv("C:/data/now_used/1015_data3.csv")
```

```python
seoul_ua = seoul[['rent_ua', '지하총층', '지상총층',
       '연면적m2', '주차대수', '승강기수', 'usage_area', 'cont_area',
       '지하철거리m', '승객수점수', '건물별전기사용량',
       '본번별평균사용량', '횡단보도거리m', '건물연차', '가까운아파트총세대수', '가까운아파트평균평당가',
       '아파트거리', '주변세대수점수', '주변평당가점수', '평균월매출금액', '평균월매출건수',
       '평균주중매출금액', '평균주말매출금액', '편의점임대료', '공시지가', '행정동생활인구']]
seoul_ua.head()
```

```python
# 주차가능여부, 승강기여부 변수 추가
lst = []
[lst.append(1) if i > 0 else lst.append(0) for i in seoul_ua['주차대수']]

seoul_ua['주차가능여부'] = lst

lst = []
[lst.append(1) if i > 0 else lst.append(0) for i in seoul_ua['승강기수']]

seoul_ua['승강기여부'] = lst

seoul_ua
```

```python
# 0값 제거
seoul_ua1 = seoul_ua.copy()

seoul_ua1 = seoul_ua1[seoul_ua1.rent_ua != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.지상총층 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.연면적m2 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.usage_area != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.cont_area != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.지하철거리m != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.승객수점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.건물별전기사용량 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.본번별평균사용량 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.횡단보도거리m != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.가까운아파트총세대수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.가까운아파트평균평당가 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.아파트거리 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.주변세대수점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.주변평당가점수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균월매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균월매출건수 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균주중매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.평균주말매출금액 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.공시지가 != 0]
seoul_ua1 = seoul_ua1[seoul_ua1.행정동생활인구 != 0]
seoul_ua1
```

```python
seoul_ua2 = seoul_ua1.copy()
seoul_ua2
```

```python
seoul_ua2_log = seoul_ua2.copy()

q1 = seoul_ua2_log["rent_ua"].quantile(0.25)
q3 = seoul_ua2_log["rent_ua"].quantile(0.75)
iqt = 6.0 * (q3 - q1)
seoul_ua2_log["rent_ua"][(seoul_ua2_log["rent_ua"] > (q3 + iqt)) | (seoul_ua2_log["rent_ua"] < (q1 - iqt))] = None

seoul_ua2_log = seoul_ua2_log.dropna(axis = 0, how= 'any')

seoul_ua2_log
```

```python
# 요약 통계량 확인하기

seoul_ua2_log = seoul_ua2_log[seoul_ua2_log['연면적m2'] != 0]
seoul_ua2_log.describe()
```

```python
seoul_ua2_log.columns
```

```python
columns = ['rent_ua', '연면적m2', 'usage_area',
       'cont_area', '지하철거리m', '승객수점수', '건물별전기사용량', '본번별평균사용량', '횡단보도거리m',
       '건물연차', '가까운아파트총세대수', '가까운아파트평균평당가', '아파트거리', '주변세대수점수', '주변평당가점수',
       '평균월매출금액', '평균월매출건수', '평균주중매출금액', '평균주말매출금액', '편의점임대료', '공시지가',
       '행정동생활인구']
```

```python
# columns에 담긴 변수 log 취하고 scaling

import math
from sklearn import preprocessing
import numpy as np

for i in columns:
    seoul_ua2_log['{}_log'.format(i)] = preprocessing.scale(np.log(seoul_ua2_log['{}'.format(i)]))
    
seoul_ua2_log
```

## RF - 변수 6개

```python
z8 = seoul_ua2_log.copy()

z8.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import pandas as pd

x = z8[['연면적m2_log',
       'usage_area_log', '승객수점수_log', '본번별평균사용량_log', '평균주말매출금액_log',
       '공시지가_log']]
y = z8['rent_ua_log']


# train/test 데이터로 나누기
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
# RandomForestRegressor 갖고오기
model = RandomForestRegressor(n_estimators = 1000, oob_score = True)
# train 데이터 학습시키기
model.fit(x_train,y_train)
# 변수들 중요도 
model.feature_importances_
# x_test(테스트)데이터로 예측결과 뽑아보기
y_pred = model.predict(x_test)

y_pred
```

```python
# rmse값 뽑기
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test,y_pred)**0.5
rmse 
```

```python
# 변수이름 + 중요도로 데이터프레임 만들기
pd.DataFrame({'변수' : ['연면적m2_log',
       'usage_area_log', '승객수점수_log', '본번별평균사용량_log', '평균주말매출금액_log',
       '공시지가_log'], 'importance' : model.feature_importances_})
```

```python
# 실제값과 예측값 비교해보기
pd.DataFrame({'y':y_test,'y_pred':y_pred})
```

```python
format(model.score(x_train, y_train))
```

```python
format(model.score(x_test, y_test))
```
